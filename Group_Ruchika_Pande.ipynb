```python
#importing libraries
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('loan.csv', low_memory=False)
```

```python
#Loading the dataset and printing it
```

```python
df.head()
```

```python
copy_of_df = df
```

```python
#Printing information of the dataset
```

```python
df.info()
```

```python
# Drop Duplicates
```

```python
df.drop_duplicates()
```

```python
# Drop rows where all the row values are NaN
```

```python
df.dropna(how='all', axis=0)
```

```python
# Drop columns where all the column values are NaN
```

```python
df.dropna(axis=1,how='all',inplace=True)

# Find percentage of NaN values in each column
round((100*df.isnull().sum())/len(df.index))
```

```python
# Drop columns in which NaN data is very high
```

```python
df.drop(labels=['next_pymnt_d','mths_since_last_record','mths_since_last_delinq'],axis=1,inplace=True)
df.info()
```

```python
#Finding out unique output values
```

```python
df['loan_status'].unique()
```

```python
# Dropping current loan status as it will not be useful
```

```python
loan_status_output = ['Fully Paid','Charged Off']
df=df[df['loan_status'].isin(loan_status_output)]
df.head()
```

```python
# Converting loan status data to numeric values for analysis
```

```python
df['loan_status']=df['loan_status'].map({'Fully Paid':0,'Charged Off':1})
df.head()
```

```python
print(df['loan_status'])
```

```python
#Extracting necessary columns
```

```python
df=df.loc[:,('loan_amnt','term','int_rate','grade','emp_length',
              'home_ownership','annual_inc','issue_d','loan_status','purpose',
             'addr_state','dti')]
df.head()
```

```python
#Derived column 'year' from the existing column 'issue_d'
```

```python
df['year'] = df['issue_d'].apply(lambda x:x.split('-')[1])
df.head()
```

```python
#Convert year type from string to int
```

```python
df['year']=df['year'].astype(int)
type(df['year'][0])
```

```python
# Countplot for analyzing number of loans taken with the increasing year
```

```python
plt.figure(figsize=(16,7))
sb.set(style="whitegrid")
ax=sb.countplot(x='year',data=df)
years = ['2007','2008','2009','2010','2011']
xpos=np.arange(len(years))
ax.set(xlabel='Years', ylabel='Total Numbers of loans taken',title='Number of Loan Claimants')
plt.xticks(xpos,years)
plt.show()
```

```python
# Analysis of the above chart
# From the above univariate analysis we see that the number of loans taken has increased with the increasing year.
```

```python
# Counterplot for analyzing purpose of loans ( Segmented univariate analysis)
```

```python
plt.figure(figsize=(16,7))
ax = sb.countplot(x='purpose',data=df,order=df['purpose'].value_counts().index)
ax.set(xlabel='Loan purpose', ylabel='No. of loan Purposes',title='Number of Loan Purpose')
plt.xticks(rotation=90)
plt.yscale('log')
plt.show()
```

```python
# From the above chart, we conclude that,
# the top 3 loan purposes are 'debt_consolidation', 'credit_card' and 'other' 
```

```python
# Distplot for loan amount distribution
```

```python
plt.figure(figsize=(16,7))
ax=sb.displot(df['loan_amnt'],bins=7,color='y')
ax.set(xlabel='Loan Amount',title='Distribution of loan amount')
plt.show()
```

```python
# It can be seen that the loan amout range is highest for 5000-10000 amount range
```

```python
# Seperate dataframe for defaulters
```

```python
defaulter_df=df.loc[df['loan_status']==1]
defaulter_df.head()
```

```python
#Calculate the percentage of loan defaulters
```

```python

grades_df=pd.DataFrame(data=round((100*defaulter_df['grade'].value_counts())/df['grade'].value_counts(),2))
grades_df.rename(columns={'grade':'defaulter_perc'},inplace=True)
grades_df
```

```python
import matplotlib.ticker as mtick
plt.figure(figsize=(16,7))
ax=sb.barplot(x=grade_df.index,y='defaulter_perc',data=grades_df)
ax.set(xlabel='Customer Grade', ylabel='Percentage of Defaulters',title='Analysis of Defaulters')
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
plt.show()
```

```python
# From the above chart we notice that the percentage of defualters is highest for G grade customers
```

```python
# Percentage of loan defaulters w.r.t to loan purpose
```

```python
purposes_df=pd.DataFrame(data=round((100*defaulter_df['purpose'].value_counts())/df['purpose'].value_counts(),2))
purposes_df.rename(columns={'purpose':'defaulter_perc'},inplace=True)
purposes_df.sort_values(by='defaulter_perc',ascending=False,inplace=True)
purposes_df
```

```python
import matplotlib.ticker as mtick
plt.figure(figsize=(16,7))
ax=sb.barplot(x=purposes_df.index,y='defaulter_perc',data=purposes_df)
ax.set(xlabel='Customer Purpose', ylabel='Percentage of Defaulters',title='Risk Analysis of Defaulters acc. to Purpose')
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
plt.xticks(rotation=90)
plt.show()
```

```python
# We see percentage is highest for small_business
```
